#Overlaping Images of Different Sizes

We can use this quick trick to quickly overlap different sized images, by simply reassigning the larger image's values to match the smaller image.

```python
# Load two images
img1=cv2.imread("/content/drive/MyDrive/Deep Learning/Computer vision/IMAGE_PROCESSING/IMAGE_PROCESSING/dog_backpack (3).png")
img2=cv2.imread("/content/drive/MyDrive/Deep Learning/Computer vision/IMAGE_PROCESSING/IMAGE_PROCESSING/watermark_no_copy (2).png")
img2 =cv2.resize(img2,(600,600))

img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

large_img = img1
small_img = img2
```

```python
large_img.shape
```

```python
small_img.shape
```

```python
plt.imshow(large_img)
```

```python
plt.imshow(small_img)
```

## top left

```python
large_img[20:620, 20:620] = small_img
```

```python
plt.imshow(large_img)
```

## special position

```python
large_img[200:200+600, 200:200+600] = small_img
```

```python
plt.imshow(large_img)
```
